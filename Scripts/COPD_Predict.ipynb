**preprocessing-Mel Spectrogram**

```python
# @title Default title text
import os
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

def compute_melspectrogram(input_path, output_path):
    # Load the audio file
    audio_binary = tf.io.read_file(input_path)
    waveform, sample_rate = tf.audio.decode_wav(audio_binary)

    # Convert waveform tensor to numpy array
    waveform_np = waveform.numpy().flatten()

    # Compute the spectrogram
    spectrogram = tf.signal.stft(waveform_np, frame_length=1024, frame_step=256)
    magnitude = tf.abs(spectrogram)

    # Convert to mel spectrogram
    num_mel_bins = 128
    num_spectrogram_bins = magnitude.shape[-1]
    upper_edge_hertz = tf.cast(sample_rate / 2, tf.float32)  # Convert to float32
    linear_to_mel_weight_matrix = tf.signal.linear_to_mel_weight_matrix(
        num_mel_bins, num_spectrogram_bins, sample_rate, 0, upper_edge_hertz
    )
    mel_spectrogram = tf.tensordot(magnitude, linear_to_mel_weight_matrix, 1)

    # Convert to decibels
    mel_spectrogram_db = tf.math.log(mel_spectrogram + 1e-6) * 10.0
    plt.figure(figsize=(10, 6))
    plt.imshow(mel_spectrogram_db.numpy(), cmap='inferno', aspect='auto', origin='lower')
    #plt.colorbar(format='%+2.0f dB')
    plt.title('Mel Spectrogram')
    plt.xlabel('Time')
    plt.ylabel('Mel Bin')
    # Save the image to the output path
    plt.savefig(output_path)

if __name__ == "__main__":
    rootdir = "/content/drive/MyDrive/MiniProjCPOD/Prediction/Res_Audio"
    outdir = "/content/drive/MyDrive/MiniProjCPOD/Prediction/Res_Image"  # Replace with your desired output directory
    # Walk through the directory and its subdirectories
    count = 0
    for root, dirs, files in os.walk(rootdir):
      for file_name in files:
        count+=1
        audio_file_path = os.path.join(root, file_name)
        out_file_path = os.path.join(outdir, str(count)+ "_mel.png")
        print(out_file_path)
        # Compute the Chromogram
        compute_melspectrogram(audio_file_path, out_file_path)
print("Done with Mespectrogram poduction")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np
import os
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# Load your custom-trained model
model = tf.keras.models.load_model('/content/drive/MyDrive/MiniProjCPOD/Saved Model/Resnet50_final_model.h5')

# Specify the directory containing the images you want to classify
image_directory = '/content/drive/MyDrive/MiniProjCPOD/Prediction/Res_Image'
count = {'COPD0':0 ,'COPD1':0 ,'COPD2':0, 'COPD3':0, 'COPD4':0}
predicted_labels = []
# List of class labels, replace with your actual labels
class_labels = ['COPD0', 'COPD1', 'COPD2', 'COPD3', 'COPD4']

# Loop through the images in the directory
for filename in os.listdir(image_directory):
    if filename.endswith(".png"):  # Adjust the file extension as needed
        img_path = os.path.join(image_directory, filename)

        # Load and preprocess the image
        img = image.load_img(img_path, target_size=(224, 224))
        img_array = image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)
        img_array = img_array / 255.0  # Normalize the pixel values

        # Make predictions
        predictions = model.predict(img_array)
        predicted_class = np.argmax(predictions, axis=1)
        predicted_label = class_labels[predicted_class[0]]
        predicted_labels.append(predicted_label)
        count[predicted_label]+=1

```

```python
confidence = {c:(count[c]/len(predicted_labels))*100 for c in count}
```
